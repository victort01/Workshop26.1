# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Numpy

```python
## Gerando 7 valores aleatórios entre 20°C e 35°C
temperaturas = np.array(np.random.randint(20,35,7))
temperaturas
```

```python
## Criando um array com os dias da semana
dias = np.array(["Domingo","Segunda","Terça","Quarta","Quinta","Sexta","Sábado"])
dias
```

# 2. Pandas

```python
## Criando um dataframe com os dias e as temperaturas
df = pd.DataFrame({"Dias":dias,"Temperaturas":temperaturas})
df
```

```python
## Calculando a temperatura média, máxima, mínima
media = df["Temperaturas"].mean()
media = media.round()
media

## Temperatura máxima
maxima = df["Temperaturas"].max()
maxima

## Temperatura mínima
minima = df["Temperaturas"].min()
minima

print(f'Temperatura média: {media}')
print(f'Temperatura máxima: {maxima}')
print(f'Temperatura mínima: {media}')
```

```python
## Extra: exibindo apenas os dias com temperatura acima da média
df_acima_media = df[df["Temperaturas"] > media]
df_acima_media
```

# 3. Matplotlib

```python
## 1. Gráfico de linhas
plt.plot(dias, temperaturas, label="Variação de temperatura ao longo da semana")
plt.ylabel('Temperatura')
plt.title('Variação de temperatura ao longo da semana')
plt.show()
```

```python
#2. Gráfico de barras
plt.bar(dias, temperaturas, label="Variação de temperatura ao longo da semana")
plt.ylabel('Temperatura')
plt.title('Variação de temperatura ao longo da semana')
plt.show()
```

```python
# 3. Gráfico de dispersão
plt.scatter(dias, temperaturas, label="Variação de temperatura ao longo da semana")
plt.ylabel('Temperatura')
plt.title('Variação de temperatura ao longo da semana')
plt.show()
```
